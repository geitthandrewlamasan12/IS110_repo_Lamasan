```python
hours = float(input("Enter Hours:"))
rate = float(input("Enter Rate:"))
if hours > 40:
  overtime = hours - 40
  pay = (40 * rate) + (overtime * rate * 1.5)
else:
  pay = hours * rate
print("Pay:", float(pay))
```

```python

Write a program to prompt for a score between 0 and 100 and print a grade using the following table:

Score           Grade
[ 97, 100]      1.0
[ 93,  97)      1.25
[ 89,  93)      1.5
[ 85,  89)      1.75
[ 81,  85)      2.0
[ 77,  81)      2.25
[ 74,  77)      2.5
[ 0,   74)      3.0
```

```python
score = float(input ("Enter your score 0 to 100: "))
if score >= 97 and score <= 100:
  print ("Grade: 1.0")
elif score >= 93 and score < 97:
         print ("Grade: 1.25")
elif score >= 89 and score < 93:
         print("Grade: 1.5")
elif score >= 85 and score < 89:
         print ("Grade: 1.75")
elif score >= 81 and score < 85:
         print("Grade: 2.0")
elif score >= 77 and score < 81:
         print ("Grade: 2.25")
elif score >= 74 and score < 77:
         print ("Grade: 2.5")
else:
         print("Grade: 3.0")
```
