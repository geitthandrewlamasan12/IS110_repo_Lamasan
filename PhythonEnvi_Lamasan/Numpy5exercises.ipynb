```python
#A.
import numpy as np

#Python nested list from 1–100
nested_list = [
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
]
A = np.array(nested_list)
print("Shape of A:", A.shape) #shows the number of rows and columns of the array
print("Maximum value using A.max():", A.max()) #Finds the maximum value in the array
print("Maximum value using np.max(A):", np.max(A))#Finds the maximum value but using a NumPy function
```

```python
#Exercise B

A = np.ones((5, 5), dtype=int)  #(5,5) means 5 rows and 5 columns; dtype=int makes them integers
print("Shape:", A.shape)  #rows, columns of the array
print("Size:", A.size)  #size returns the total number of elements in the array
print("Number of dimensions:", A.ndim)
```

```python
#Exercise C

A = np.arange(-5, 6)  # np.arange(start, stop) creates values from -5 up to 5 (6 is excluded)
result = A * 2  # This multiplies every element in the array by 2
print("Resulting array:", result)  # Displays the new array after multiplication
```
