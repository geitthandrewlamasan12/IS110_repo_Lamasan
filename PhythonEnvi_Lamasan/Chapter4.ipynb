```python
Exercise 1
Rewrite your pay computation with time-and-a-half for overtime and create a function called compute pay which takes two parameters (hours and rate).

Enter Hours: 45
Enter Rate: 10
Pay: 475.0
```

```python
def compute_pay(hrs, rte):
  if hrs > 40:
    overtime = hrs - 40
    pay = float(40 * rte) + float(overtime * rte * 1.5)
  else:
    pay = hrs * rte
  return pay

hrs = float(input ("Enter Hours: "))
rte = float(input ("Enter Rate: "))

result = compute_pay(hrs, rte)
print("Pay: ", result)
```

```python
CHALLENGE
Challenge
Write a function called computegrade that takes a score as its parameter and returns a grade as a string using the following table:

Score           Grade
[ 97, 100]      1.0
[ 93,  97)      1.25
[ 89,  93)      1.5
[ 85,  89)      1.75
[ 81,  85)      2.0
[ 77,  81)      2.25
[ 74,  77)      2.5
[ 0,   74)      3.0

Run the program repeatedly to test the various different values for input.
```

```python
def computergrade(score):
  if score >= 97 and score <= 100:
    return "1.0"
  elif score >= 93 and score < 97:
    return "1.25"
  elif score >= 89 and score < 93:
    return "1.5"
  elif score >= 85 and score < 89:
    return "1.75"
  elif score >= 81 and score < 85:
    return "2.0"
  elif score >= 77 and score < 81:
    return "2.25"
  elif score >= 74 and score < 77:
    return "2.5"
  elif score >= 0 and score < 74:
    return "3.0"
  else:
    return "Score Grade Invalid"

while True:
  users_input = input ("Enter score or type 'q' to quit:")

  if users_input.lower() == 'q':
    break

  score = float(users_input)
  grade = computergrade(score)
  print("Grade: ", grade)
```
